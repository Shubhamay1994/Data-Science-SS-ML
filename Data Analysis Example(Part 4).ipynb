    Return value from get():
    1.the value for the specified  key if 'key' is in dictonary
    2.None if the 'key' is not not found and value is not specified
    3.value of the key is not found and value is specified

```python
person={'name':'Phill','age':22}
#Key is in the dictionary
print('Name: {}, Age:{}'.format(person.get('name'),person.get('age')))
#value is not provided
print('Salary: ',person.get('salary'))
#if value is provide
print('Salary: ',person.get('salary',100))
```

```python
#You will notice by lookingat the oocupation many have the same basic occupation.so,here mapping is required
occ_mapping={'INFORMATION REQUESTED PER BEST EFFORTS' : 'NOT PROVIDED',\
             'INFORMATION REQUESTED' : 'NOT PROVIDED',\
             'INFORMATION REQUESTED (BEST EFFORTS)' : 'NOT PROVIDED',\
             'C.E.O.': 'CEO'}
```

```python
f=lambda x:occ_mapping.get(x,x)
fec.contbr_occupation=fec.contbr_occupation.map(f)
```

```python
emp_mapping={'INFORMATION REQUESTED PER BEST EFFORTS' : 'NOT PROVIDED','INFORMATION REQUESTED' : 'NOT PROVIDED',\
             'SELF' : 'SELF-EMPLOYED','SELF EMPLOYED' : 'SELF-EMPLOYED',}
```

```python
f=lambda x:emp_mapping.get(x,x)
fec.contbr_occupation=fec.contbr_occupation.map(f)
```

```python
fec['contbr_occupation'].value_counts()[:10]
```

```python
#now you can use pivot table to aggregate the data by occupation and party then filterdown to the subset that denoted atleast $2 million overall
by_occupation=fec.pivot_table('contb_receipt_amt',index='contbr_occupation',columns='party',aggfunc='sum')
by_occupation
```

```python
by_occupation.sum(1)
```

```python
over_2mm=by_occupation[by_occupation.sum(1)>2000000]
```

```python
over_2mm
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,8)
over_2mm.plot.barh()
```

```python
#tou might be interest to know about the yop occupation and top company name that donated to obama and romey
#to do this you can group by the candidate name and use varient of the top method
def get_top(group,key,n=5):
    total=group.groupby(key)['contb_receipt_amt'].sum()
    return total.nlargest(n)
```

```python
grouped=fec_rmob.groupby('cand_nm')
grouped.apply(get_top,'contbr_occupation',n=7)
```

```python
grouped.apply(get_top,'contbr_employer',n=10)
```

    Bucketing Donation Amount

```python
import numpy as np
bins=np.array([0,1,10,100,1000,10000,100000,1000000,10000000])
```

```python
label=pd.cut(fec_rmob.contb_receipt_amt,bins)
```

```python
label
```

```python
grouped=fec_rmob.groupby(['cand_nm',label])
```

```python
grouped.size()
```

```python
#from the data we can observe that obama get larger number ofsmall donation than Romney
grouped.size().unstack(0)
```

```python
bucket_sum=grouped.contb_receipt_amt.sum().unstack(0)
```

```python
bucket_sum
```

```python
bucket_sum.sum(axis=1)
```

```python
bucket_sum.sum(axis=0)
```

```python
normed_sums=bucket_sum.div(bucket_sum.sum(axis=1),axis=0)
```

```python
normed_sums
```

```python
normed_sums[:-2].plot.barh()
```
