# Qantile and Bucket analysis:

```python
import pandas as pd
import numpy as np
frame=pd.DataFrame({'data1':np.random.rand(1000),'data2':np.random.rand(1000)})
frame
```

```python
frame.max()
```

```python
frame.min()
```

```python
Quartile=pd.cut(frame.data1,4)
```

```python
Quartile[:10]
```

```python
def get_status(group):
    return{'min':group.min(),'max':group.max(),'count':group.count(),'mean':'mean'}
```

```python
grouped=frame.data2.groupby(Quartile)
grouped.apply(get_status)
```

```python
grouped.apply(get_status).unstack()
```

```python
frame.data2.describe()
```

```python
pd.cut(frame.data1,10).value_counts()
```

qcut:Discretize variable into equal-sized buckets based on rank or based on sample quantiles. For example 1000 values for 10 quantiles would produce a Categorical object indicating quantile membership for each data point.

```python
grouped=pd.qcut(frame.data2,10)
grouping=frame.data2.groupby(grouped)
grouping.apply(get_status).unstack()
```

```python
grouped=pd.qcut(frame.data2,10,labels=False)
grouping=frame.data2.groupby(grouped)
grouping.apply(get_status).unstack()
```

```python
#construct a deck of playing card
suit=['H','S','C','D']
card_val=(list(range(1,11))+[10]*3)*4
base_name=['A']+list(range(2,11))+['J','K','Q']
cards=[]
for card in suit:
      cards.extend(str(num)+card for num in base_name)
deck=pd.Series(card_val,index=cards)
```

```python
def draw(deck,n=5):
    return deck.sample(n)
```

```python
draw(deck)
```

```python
get_suit=lambda card:card[-1]
```

```python
deck.groupby(get_suit).apply(draw,n=3)
```

```python
deck.groupby(get_suit,group_keys=False).apply(draw,n=3)
```

# Group Wise Linear Regression

```python
import statsmodels.api as sm
```

```python
#full form of ols means-->ordinary least squre
# a simple ordinary least squre model
```

In line with that, OLS is an estimator in which the values of m and c (from the above equation) are chosen in such a way as to minimize the sum of the squares of the differences between the observed dependent variable and predicted dependent variable. Thats why its named ordinary least squares.

```python
#in line with that ,OLS is an estimator in which the alue of m and c (from)
```

```python
def regress(data,yvar,xvar):
    Y=data[yvar]
    X=data[xvar]
    X['intercept']=1
    result=sm.OLS(Y,X).fit()
    return result.params
```

```python
by_year.apply(regress,'AAPL',['SPX'])
```
