# Merging DataFrame

```python
import pandas as pd
employee_data=pd.DataFrame({'employee_id':['1','2','3','4'],'name':['Amy john','Ram Biswas','putul Rani','mana bul']})
```

```python
employee_data
```

```python
employee_sales=pd.DataFrame({'employee_id':['3','4','5','6'],'sales':[2200,2000,5000,3500]})
employee_sales
```

```python
#by default pd.merge compute inner join
pd.merge(employee_data,employee_sales,on='employee_id')
```

```python
pd.merge(employee_data,employee_sales,how='outer',on='employee_id')
```

```python
pd.merge(employee_data,employee_sales,how='left',on='employee_id')
```

```python
l=pd.DataFrame({'key1':['a','b','c','d','e'],'key2':['g','h','i','j','k'],'value':[1,2,3,4,5]})
l
```

```python
g=pd.DataFrame({'key1':['aa','bb','cv','dd','ee'],'key2':['gg','hh','ii','jj','kk'],'value':[11,21,31,41,51]})
g
```

```python
#join with two key
pd.merge(l,g,on=['key1','key2'],how='outer')
```

```python
#join with an one key
pd.merge(l,g,on=['key1'],how='outer')
```

```python
pd.merge(l,g,left_on='key1',right_on='key2',how='outer')
```
