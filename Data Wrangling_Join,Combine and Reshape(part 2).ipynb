# Pivoting "Wide" to "Long" Format:

```python
df=pd.DataFrame({'key':['foo','bar','baz'],'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
df
```

```python
df1=pd.melt(df,['key'])
df1
```

```python
reshape=df1.pivot('key','variable','value')
reshape
```

```python
df1.reset_index()
```

```python
df.reset_index()
```

```python
pd.melt(df,value_vars=['A','B','C'])
```

```python
pd.melt(df,id_vars='key',value_vars=['A','B','C'])
```
