```python
#to annotated some feature of the plot we use,
#.annotate(xy,xytext,arrowprp)
#xy-->the location being annotated
#xytext-->the location of the text
#arrowprop=draws an arrow from the text to the annoted point by giving a dictionary of arrow properties 
```

```python
from datetime import datetime
data=pd.read_csv(r'C:\Users\Shubhamay\Downloads\pydata-book-master_ch08_spx.csv',index_col=0,parse_dates=True)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
spx=data['SPX']
spx.plot(ax=ax,color='r')
crisis_date=[(datetime(2007,10,11),'Peak bull market'),\
            (datetime(2008,3,12),'Bear stern fails'),\
            (datetime(2008,9,15),'Lehaman Bankrupcy')]
for date,level in crisis_date:
    ax.annotate(level,xy=(date,spx.asof(date)+75),xytext=(date,spx.asof(date)+180),\
               arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4))
ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim(600,1800)
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

# Drawing shape in matplotlib

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
rec=plt.Rectangle((0.2,0.75),0.15,0.18,color='red')
circle=plt.Circle((0.2,0.15),0.1,color='yellow')
polygon=plt.Polygon([[0.7,0.8],[0.8,0.9],[0.9,0.8]],color='green')
ax.add_patch(rec)
ax.add_patch(circle)
ax.add_patch(polygon)
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
df=pd.read_csv(r'C:\Users\Shubhamay\Downloads\pydata-book-master_ch07_macrodata.csv')
df
```

```python
#diff means 0-1
data=df[['cpi','m1','tbilrate','unemp']]
trans_data=np.log(data).diff().dropna()
trans_data
```

```python
plt.rc('figure',figsize=(10,8))
plt.scatter(trans_data['m1'],trans_data['unemp'])
```

```python
##seaborn scatter plot with linear regression line
sns.regplot('m1','unemp',data=trans_data,color='r')
plt.title('change in log %s vercus log %s' %('m1','unemp'))
```

    pair plot

```python
#KDE-->Kernal Density Estimation,is way to estimate probability density 
#function of a continious random variable . it is used to \
# do non_parametric analysis
sns.pairplot(trans_data,diag_kind='kde',plot_kws={'alpha':0.2})
```

```python
trans_data['cpi'].plot.density()
```

```python
sns.pairplot(trans_data,diag_kind='hist',plot_kws={'alpha':0.2})
```
